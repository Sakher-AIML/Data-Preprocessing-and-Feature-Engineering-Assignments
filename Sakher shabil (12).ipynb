```python
import numpy as np
from scipy.stats import norm

# Select the column
data = df['SepalLengthCm']

# Calculate mean and standard deviation
mean = np.mean(data)
sd = np.std(data)

# Generate x values for smooth curve
x = np.linspace(min(data), max(data), 200)

# Plot histogram
plt.figure()
plt.hist(data, bins=20, density=True, alpha=0.6)

# Plot normal distribution curve
plt.plot(x, norm.pdf(x, mean, sd), linewidth=2)

plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.ylabel("Density")
plt.show()
```

```python
# Select the column
data = df['SepalWidthCm']

# Calculate mean and standard deviation
mean = np.mean(data)
sd = np.std(data)

# Generate x values for smooth curve
x = np.linspace(min(data), max(data), 200)

# Plot histogram
plt.figure()
plt.hist(data, bins=20, density=True, alpha=0.6)

# Plot normal distribution curve
plt.plot(x, norm.pdf(x, mean, sd), linewidth=2)

plt.title("Sepal Width Distribution")
plt.xlabel("Sepal Width")
plt.ylabel("Density")
plt.show()
```

```python
# Plot histogram with KDE curve
plt.figure()
sns.histplot(df['PetalLengthCm'],kde=True,color='blue', bins=20)
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Density")
plt.show()
```

```python
# Plot histogram with KDE curve
plt.figure()
sns.histplot(df['PetalWidthCm'], kde=True,color='blue', bins=20)
plt.title("Petal Width Distribution")
plt.xlabel("Petal Width")
plt.ylabel("Density")
plt.show()
```

```python
# Count frequency of each species
counts = df['Species'].value_counts()
# Plot bar chart
plt.figure()
counts.plot(kind='bar')
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

```python
# Count frequency of each species
counts = df['Species'].value_counts()

# Plot pie chart showing species distribution
plt.figure()
counts.plot(kind='pie', autopct='%1.1f%%')

plt.title("Species Distribution")
plt.ylabel("")
plt.show()
```
