```python
df['Price_per_sqft'] = df['price'] / df['area']
```

```python
def area_category(x):
    if x < 1000:
        return 'Small'
    elif 1000 <= x <= 2000:
        return 'Medium'
    else:
        return 'Large'

df['Area_Category'] = df['area'].apply(area_category)
```

```python
df[['area', 'Area_Category']].head()
```

```python
df['Price_Rank'] = df['price'].rank(ascending=False)
```

```python
df['Area_Rank'] = df['area'].rank(ascending=False)
```

```python
df['Rank_within_furnish'] = df.groupby('furnishingstatus')['price'].rank(ascending=False)
```

```python
house_price = df['price'].quantile(0.95)
df[df['price'] >= house_price]
```

```python
Q1 = df['Price_per_sqft'].quantile(0.25)
Q3 = df['Price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR

df[df['Price_per_sqft'] > upper]
```
