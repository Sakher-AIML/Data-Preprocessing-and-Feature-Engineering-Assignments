```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/tips (1).csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 1. Total Bill vs Tip (Scatter)
# Shows relationship between total bill and tip amount
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.title('total bill vs tip')
plt.show()
```

```python
# 2. Total Bill vs Tip (Regression)
# Shows trend line between total bill and tip
sns.regplot(x='total_bill',y='tip',data=df)
plt.title('total bill vs tip')
plt.show()
```

```python
# 3. Day vs Total Bill (Box Plot)
# Compares total bill distribution across days
sns.boxplot(x='day',y='total_bill',data=df)
plt.title('day vs total bill')
plt.show()
```

```python
# 4. Day vs Tip (Box Plot)
# Compares tip distribution across different days
sns.boxplot(x='day',y='tip',data=df)
plt.title('day vs tip')
plt.show()
```

```python
# 5. Smoker vs Tip (Box Plot)
# Compares tip amounts between smokers and non-smokers
sns.boxplot(x='smoker',y='tip',data=df)
plt.title('smoker vs tip')
plt.show()
```

```python
# 6. Time vs Total Bill (Box Plot)
# Compares total bill between lunch and dinner time
sns.boxplot(x='time',y='total_bill',data=df)
plt.title('time vs total bill')
plt.show()
```

```python
# 7. Size vs Total Bill (Bar Plot)
# Shows how total bill changes with group size
sns.barplot(x='size',y='total_bill',data=df)
plt.title('size vs total bill')
plt.show()
```

```python
# 8. Size vs Tip (Scatter)
# Shows relationship between group size and tip amount
sns.scatterplot(x='size',y='tip',data=df)
plt.title('size vs tip')
plt.show()
```

```python
# 9. Total Bill vs Tip (Hue = Smoker)
# Compares tipping behavior for smokers vs non-smokers
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=df)
plt.title('total bill vs tip')
plt.show()
```

```python
# 10. Day vs Average Total Bill (Line Plot)
# Shows how the average total bill changes across days
sns.lineplot(x='day',y='total_bill',data=df)
plt.title('day vs  Average total bill')
plt.show()
```

```python
# 11. Day vs Average Tip (Line Plot)
# Shows how the average tip changes across days
sns.lineplot(x='day',y='tip',data=df)
plt.title('day vs  Average tip')
plt.show()
```

```python
# Shows correlation between variables, where darker colors indicate stronger relationships
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
