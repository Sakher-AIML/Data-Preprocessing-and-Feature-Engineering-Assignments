```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/tips (1) (1).csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#check null values
df.isnull().sum()
```

```python
#check duplicate
df.duplicated().sum()
```

```python
#Remove duplicate
df.drop_duplicates(inplace=True)
```

```python
#check duplicate
df.duplicated().sum()
```

```python
# Boxplot shows distribution of total_bill across different days
sns.boxplot(x='day', y='total_bill', hue='sex', data=df)
plt.title('Day vs Total Bill')
plt.show()
```

```python
# Boxplot shows distribution of total_bill across different time
sns.boxplot(x='time', y='total_bill', hue='sex', data=df)
plt.title('Time vs Total Bill')
plt.show()
```

```python
# Barplot shows average (mean) total_bill vs smoker
sns.barplot(x='smoker', y='total_bill', hue='sex', data=df)
plt.title('Day vs Total Bill')
plt.show()
```

```python
# Barplot shows average (mean) total_bill vs day
sns.barplot(x='day', y='total_bill', hue='sex', data=df)
plt.title('Day vs Total Bill')
plt.show()
```

```python
# Line plot shows trend of day vs total_bill
sns.lineplot(x='day', y='total_bill', hue='sex', data=df)
plt.title('Day vs Total Bill')
plt.show()
```

```python
# Line plot shows trend of day vs tip
sns.lineplot(x='day', y='tip', hue='time', data=df)
plt.title('Day vs Tip')
plt.show()
```

```python
# Scatter plot shows relationship between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# Scatter plot shows relationship between size and tip
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.title('Size vs Tip')
plt.show()
```

```python
# Pairplot shows relationships between all numerical variables using smoker as hue
sns.pairplot(df[['total_bill','tip','size','smoker']],hue='smoker')
plt.show()
```

```python
# Pairplot shows relationships between all numerical variables using sex as hue
sns.pairplot(df[['total_bill','tip','size','sex']],hue='sex')
plt.show()
```

```python
# Heatmap shows correlation between numerical variables
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
