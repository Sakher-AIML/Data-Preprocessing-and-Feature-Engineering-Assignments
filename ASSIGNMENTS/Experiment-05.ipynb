```python
def mean_or_median(series):
  Q1=series.quantile(0.25)
  Q3=series.quantile(0.75)
  IQR=Q3-Q1
  has_outlies=(series<Q1-1.5*IQR) | (series>Q3+1.5*IQR)
  return "median"if has_outlies.any() else "mean"
for col in numarical_cols:
  print(col,'→',mean_or_median(df[col]))
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df.isnull().sum()
```

```python
#create new Titel column
#extract titles(e.g,Mr,Mrs,Dr)from the 'Name' column and stores them in a new 'Title' column
df['Title']=df['Name'].str.extract(r',\s*([A-Za-z]+)\.?',expand=True)
#counts the frequency of each extracted title
df['Title'].value_counts()
```

```python
mapping = {
    'Mr': 'Mr',
    'Miss': 'Miss',
    'Mrs': 'Mrs',
    'Master': 'Master',
    'Dr': 'Rare',
    'Rev': 'Rare',
    'Col': 'Rare',
    'Mlle': 'Miss',
    'Major': 'Rare',
    'Ms': 'Miss',
    'Mme': 'Mrs',
    'Don': 'Rare',
    'Lady': 'Rare',
    'Sir': 'Rare',
    'Capt': 'Rare',
    'the': 'Rare',
    'Jonkheer': 'Rare'
}
df.replace({'Title': mapping}, inplace=True)
df['Title'].value_counts()
```
