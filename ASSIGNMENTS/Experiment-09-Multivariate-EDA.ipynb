```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/penguins.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['bill_depth_mm']=df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean())
df['bill_depth_mm']=df['bill_depth_mm'].fillna(df['bill_depth_mm'].mean())
df['flipper_length_mm']=df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean())
df['body_mass_g']=df['body_mass_g'].fillna(df['body_mass_g'].mean())
df['sex']=df['sex'].fillna(df['sex'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
#Boxplot shows distribution of bill_length_mm across different species
sns.boxplot(x='species',y='bill_length_mm',hue='sex',data=df)
plt.title('Species vs Bill Length')
plt.show()
```

```python
#Boxplot shows distribution of bill_depth_mm across different island
sns.boxplot(x='island',y='bill_depth_mm',hue='sex',data=df)
plt.title('Island vs Bill Depth')
plt.show()
```

```python
# Barplot shows average (mean) species vs body_mass_g
sns.barplot(x='species',y='body_mass_g',hue='sex',data=df)
plt.title('Species vs Body Mass')
plt.show()
```

```python
# Barplot shows average (mean) island vs body_mass_g
sns.barplot(x='island',y='body_mass_g',hue='sex',data=df)
plt.title('Island vs Body Mass')
plt.show()
```

```python
# Scatter plot shows relationship between bill length and bill depth
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',hue='species',data=df)
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
# Scatter plot shows relationship between flipper_length and body mass_g
sns.scatterplot(x='flipper_length_mm',y='body_mass_g',hue='island',data=df)
plt.title('Flipper Length vs Body Mass')
plt.show()
```

```python
# Pairplot shows relationships between all numerical variables using sex as hue
sns.pairplot(df[['bill_length_mm','bill_depth_mm',	'flipper_length_mm','sex']],hue='sex')
plt.show()
```

```python
# Pairplot shows relationships between all numerical variables using species as hue
sns.pairplot(df[['bill_length_mm','bill_depth_mm',	'flipper_length_mm','species']],hue='species')
plt.show()
```

```python
# Heatmap shows correlation between numerical variables
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
