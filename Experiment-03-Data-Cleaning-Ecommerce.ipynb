```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/indian_ecommerce_orders.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Quantity'] = df['Quantity'].replace({'zero':0})
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
df['CustomerRating'] = df['CustomerRating'].replace({'one':1,'two':2,'three':3,'four':4,'five':5})
df['CustomerRating'] = pd.to_numeric(df['CustomerRating'], errors='coerce')
```

```python
df['DiscountPercent'] = df['DiscountPercent'].replace({'ten%':10})
df['DiscountPercent'] = df['DiscountPercent'].str.replace('%','')
df['DiscountPercent'] = pd.to_numeric(df['DiscountPercent'], errors='coerce')
```

```python
df['OrderAmount'] = pd.to_numeric(df['OrderAmount'], errors='coerce')
df['ShippingFee'] = pd.to_numeric(df['ShippingFee'], errors='coerce')
df['DeliveryDays'] = pd.to_numeric(df['DeliveryDays'], errors='coerce')
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['CustomerRating'], inplace=True)
```

```python
numarical_cols=df.select_dtypes(include=['int64','float64']).columns
catecorical_cols=df.select_dtypes(include=['object','category']).columns
print('Numerical_cols:',list(numarical_cols))
print('catecorical_cols:',list(catecorical_cols))
```

```python
df['Quantity']=df['Quantity'].fillna(df['Quantity'].mean())
df['OrderAmount']=df['OrderAmount'].fillna(df['OrderAmount'].mean())
df['DiscountPercent']=df['DiscountPercent'].fillna(df['DiscountPercent'].mean())
df['ShippingFee']=df['ShippingFee'].fillna(df['ShippingFee'].mean())
df['DeliveryDays']=df['DeliveryDays'].fillna(df['DeliveryDays'].mean())
df['OrderDate']=df['OrderDate'].fillna(df['OrderDate'].mean())
```

```python
df['Pincode']=df['Pincode'].fillna(df['Pincode'].mode()[0])
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Returned']=df['Returned'].fillna(df['Returned'].mode()[0])
```
